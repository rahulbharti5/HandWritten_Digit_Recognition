```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
data = pd.read_csv('Book1.csv')

# Plot a line graph
plt.plot(data['SC'], data['SEAT'])
plt.xlabel('Category cut off')
plt.ylabel('Number of Seats')
plt.title('Line Graph')
plt.show()

# Plot a scatter plot
plt.scatter(data['SC'], data['SEAT'])
plt.xlabel('Category cut off')
plt.ylabel('Number of Seats')
plt.title('Scatter Plot')
plt.show()

# Plot a line graph
plt.plot(data['SEAT'], data['SC'])
plt.xlabel('Number of Seats')
plt.ylabel('Category cut off')
plt.title('Line Graph')
plt.show()

# Plot a scatter plot
plt.scatter(data['SEAT'], data['SC'])
plt.xlabel('Number of Seats')
plt.ylabel('Category cut off')
plt.title('Scatter Plot')
plt.show()
```
